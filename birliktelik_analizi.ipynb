```python
#gerekli kütüphaneleri yüklüyoruz
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
#ilgili datasetini yüklüyoruz, başlığı products olarak değiştiriyoruz.
df=pd.read_csv("GroceryStoreDataset.csv" , names=["products"],header=None)
df.head()
```

```python
#dataseti boyutuna bakıyoruz
df.shape
```

```python
#sütunu kontrol ediyoruz.
df.columns
```

```python
#ürünlere göz atıyoruz
df.values
```

```python
#lambda ile producst altındaki ürünleri virgülden sonra olacak şekilde ayırıyoruz
data=list(df["products"].apply(lambda x: x.split(",")))
data
```

```python
#encoder işlemi için TransactionEncoder'ı yüklüyoruz
from mlxtend.preprocessing import TransactionEncoder
```

```python
#Encoder işlemi ile satırlara göre elimizdeki ürünlerin satın alınıp alınmadığını True False olarak çeviriyoruz
te = TransactionEncoder()
te_data = te.fit(data).transform(data)
df = pd.DataFrame(te_data,columns=te.columns_)
df
```

```python
#Apriori algoritmasına göre ürünlerin support değerlerini çıkarıyoruz
df1 = apriori(df,min_support=0.2,use_colnames=True,verbose = 1)
df1
```

```python
#support değerlerine göre büyükten küçüğe sıralama yapıyoruz.
df1.sort_values(by="support",ascending=False)
```

```python
#association_rules ile confidence değerini getiriyoruz.
association_rules(df1, metric="confidence", min_threshold = 0.6)
```

```python
#confidence değerinin 0,4'ten büyük olanlarını çağırıyoruz.
df_ar = association_rules(df1, metric = "confidence", min_threshold = 0.4)
df_ar
```

```python
#support değeri 0,6'dan küçük olan, confidence değerinin 0,3'ten büyük olan değerlerini çağırıyoruz
df_ar[(df_ar.support < 0.6) & (df_ar.confidence > 0.3)]
```
